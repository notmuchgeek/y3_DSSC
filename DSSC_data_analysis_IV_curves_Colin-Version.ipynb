# Z907

```python
figure = figsize(12,8)
z907dark=genfromtxt('all_plot_datasets\SC1_Z907_dark.txt',delimiter=' ')
z907light=genfromtxt('all_plot_datasets\SC1_Z907_lighted.txt',delimiter=' ')
vz907dark=z907dark[:,0]
iz907dark=z907dark[:,1]
vz907light=z907light[:,0]
iz907light=z907light[:,1]
plot(vz907dark,iz907dark,label = 'dark')
plot(vz907light,iz907light,label = 'light')
plot(0+vz907dark-vz907dark,-0.000595059+iz907dark-iz907dark,marker='x',label='I(SC)= -0.595059 mA')
plot(0.435+vz907dark-vz907dark,0+iz907dark-iz907dark,marker='x',label='V(OC)= 0.435 V')
plot(vz907dark,0+iz907dark-iz907dark,linestyle='--')
title('I/V curves for cell with Z907-dyed electrode')
xlabel('Voltage/V')
ylabel('Current/A')
legend()
```
