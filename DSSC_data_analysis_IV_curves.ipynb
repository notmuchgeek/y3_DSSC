```python
%pylab inline
import pandas as pd
from scipy.signal import find_peaks
import plotly.graph_objects as go

import matplotlib.pyplot as plt
import numpy as np
```

```python
import os

#Universal path for the working dictionary
PATH = 'C:/Users/AlbertCielstian/Documents/GitHub/y3_DSSC/all_plot_datasets/'

files = os.listdir(PATH)

for file in files:
    print(file)
```

```python
def load_d(name):
    xy_data=np.loadtxt(PATH+str(name),skiprows=0)

    x1=[]
    y1=[]
    for i in xy_data:
        x1.append(i[0])
        y1.append(i[1])
    return np.array(x1), np.array(y1)
```

```python
x_Z907_dk, y_Z907_dk = load_d('SC1_Z907_dark.txt')
x_Z907_lgt, y_Z907_lgt = load_d('SC1_Z907_lighted.txt')

x_Tur_dk, y_Tur_dk = load_d('SC3_Tur_dark.txt')
x_Tur_lgt, y_Tur_lgt = load_d('SC3_Tur_med-light.txt')

x_Beet_dk, y_Beet_dk = load_d('SC4_Beet_dark.txt')
x_Beet_lgt, y_Beet_lgt = load_d('SC4_Beet_med-light.txt')

x_Blue_dk, y_Blue_dk = load_d('SC5_Blue.txt')
x_Blue_lgt, y_Blue_lgt = load_d('SC5_Blue_med-light.txt') # not using SC5_Blue_large-light.txt here
```

```python
a,b,p,q=map(float,input('Enter the coordinates of the points:').split())

m=(q-b)/(p-a)
y=b
x=a
c=y-(m*x)

#to find x-intercept.
y=0
x=(y-c)/m
print('x-intercept of the line:',x)

'''
#to find y-intercept.
x=0
y=(m*x)+c
print('y-intercept of the line:',y)'''
```

# Plot all the I-V Curve

```python
fig, ax = plt.subplots()
ax.plot(x_Z907_dk, y_Z907_dk, linewidth=1,label = 'Z907 dark', marker='.')
ax.plot(x_Z907_lgt, y_Z907_lgt, linewidth=1,label = 'Z907 lighted', marker='.')

ax.plot(0,-0.000595059,marker='x',label='I(SC)= -0.595059 mA')
ax.plot(0.435,0,marker='x',label='V(OC)= 0.435 V')

ax.set_xlabel('Voltage (V)')
ax.set_ylabel('Current (A)')

ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.title('I/V curves for Cell with Z907-dyed Electrode')


ax.legend()
```
