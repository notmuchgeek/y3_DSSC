```python
%pylab inline
import pandas as pd
from scipy.signal import find_peaks
import plotly.graph_objects as go

import matplotlib.pyplot as plt
import numpy as np
```

```python
import os

#Universal path for the working dictionary
PATH = 'C:/Users/AlbertCielstian/Documents/GitHub/y3_DSSC/all_plot_datasets/'

files = os.listdir(PATH)

for file in files:
    print(file)
```

```python
def load_d(name):
    xy_data=np.loadtxt(PATH+str(name),skiprows=0)

    x1=[]
    y1=[]
    for i in xy_data:
        x1.append(i[0])
        y1.append(i[1])
    return np.array(x1), np.array(y1)
```

```python
x_Z907_dk, y_Z907_dk = load_d('SC1_Z907_dark.txt')
x_Z907_lgt, y_Z907_lgt = load_d('SC1_Z907_lighted.txt')

x_Tur_dk, y_Tur_dk = load_d('SC3_Tur_dark.txt')
x_Tur_lgt, y_Tur_lgt = load_d('SC3_Tur_med-light.txt')

x_Beet_dk, y_Beet_dk = load_d('SC4_Beet_dark.txt')
x_Beet_lgt, y_Beet_lgt = load_d('SC4_Beet_med-light.txt')

x_Blue_dk, y_Blue_dk = load_d('SC5_Blue.txt')
x_Blue_lgt, y_Blue_lgt = load_d('SC5_Blue_med-light.txt') # not using SC5_Blue_large-light.txt here
```

```python
fig, ax = plt.subplots()
ax.plot(x_Z907_dk, y_Z907_dk, linewidth=3,label = 'Z907 dark')
ax.plot(x_Z907_lgt, y_Z907_lgt, linewidth=3,label = 'Z907 lighted')

ax.set_xlabel('Current (A)')
ax.set_ylabel('Voltage (V)')

ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.legend()
```
