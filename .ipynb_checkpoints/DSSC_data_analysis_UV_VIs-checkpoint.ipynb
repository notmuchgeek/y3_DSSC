# Plot absorption spectrum

```python
def load_d(name):
    xy_data=np.loadtxt(PATH+str(name),skiprows=4,delimiter='\t',dtype=str)

    x1=[]
    y1=[]
    for i in xy_data:
        x1.append(i[0])
        y1.append(i[1])
        
    x_float = []
    y_float = []
    
    for m in y1:
        if m == 'XXX.XXX':
            y_float.append(0)
        else:
            y_float.append(float(m))
        
    for m in x1:
        x_float.append(float(m))

    return np.array(x_float), np.array(y_float)


def normal(a):
    amin, amax = min(a), max(a)
    for i, val in enumerate(a):
        a[i] = (val-amin) / (amax-amin)
    return np.array(a)
```
